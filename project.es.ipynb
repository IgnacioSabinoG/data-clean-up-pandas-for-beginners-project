#### Ejercicio 04. ¿Cuantas poblaciones contiene el dataset? (★☆☆)

Cuenta el número de poblaciones únicas en la columna 'level5' e imprime el nombre de las poblaciones separadas por coma. Por ejemplo:

`> print(populations)`

`population1, population2, population3, ...`

```python
populations = ds['level5'].tolist()
u_populations = set(populations)
popu_no_null = []
for i in u_populations:
    if pd.notna(i): popu_no_null.append(i)
print(", ".join(str(x) for x in popu_no_null))
```

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
# Trazar el histograma de precios
pr_am = ds.loc[ds['level5'].str.contains('Arroyomolinos (Madrid)', case=False, na=False, regex=False), 'price']

plt.figure(figsize=(8, 5))  # tamaño del gráfico
plt.hist(pr_am, bins=20, edgecolor='black')

plt.title('Distribución de precios en Arroyomolinos')
plt.xlabel('Precio (€)')
plt.ylabel('Número de viviendas')

media = pr_am.mean()
plt.axvline(media, color='red', linestyle='--', linewidth=2, label=f'Media = {media:.0f} €')
plt.legend()
plt.show()

plt.savefig("hist_precios_arroyo.png", dpi=300)
plt.show()
print("En Arroyomolinos el promedio es de unos 300 mil $, variendo entre 200 y 400 mil mayormente.")
```

El promedio de precio de las propiedades en Arroyomolinos (Madrid), es de aproximadamente 295.000 euros, aunque muestra mucha dispersión con cantidad de casas entre 200.000 euros y 380.000 euros.
